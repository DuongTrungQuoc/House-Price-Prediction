 ### 1. Problem Definition (Định nghĩa vấn đề, cho mọi người biết cái project này được dùng để làm gì):
 - Mục tiêu: Dự đoán giá bán của từng căn nhà. Mỗi căn nhà có id và các thuộc tính, dựa vào đó để dự đoán giá bán sao cho gần và chính xác nhất so với giá thực tế.
 ### 2. Feature Selection (Chọn đặc trưng để huấn luyện mô hình máy học):
 - Chọn một vài đặc trưng (cột) trong số các đặc trưng có sẵn mà bạn nghĩ tốt nhất để cho mô hình này dự đoán một cách chính xác nhất
 ### 3. Spliting The Datasets
 #### 3.1 dataset -> X,Y
 - `data` = dataset
 - `X` = `data[features]`
 - `Y` = target variable `SalePrices`
 #### 3.2 X,Y -> X_train, Y_train, X_valid, Y_valid
 ### 4. Training Machine Learning Model

## 2.Feature Selection

```python
features = ["LotArea","YearBuilt","1stFlrSF","2ndFlrSF","FullBath","BedroomAbvGr","TotRmsAbvGrd"]
```

## 3.Spliting Dataset into X && Y

```python
X = data[features]
Y = data["SalePrice"]
```

```python
X.head()
```

```python
Y.head()
```

## X,Y -> X_train, Y_train, X_valid, Y_valid

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, Y_train, Y_valid = train_test_split(X,Y,train_size= 0.8, test_size= 0.2, random_state=0)
```

```python
X_train
```

```python
X_train.shape
```

```python
X.shape
```

```python
X_valid.shape
```

 ## 4. Training Machine Learning Model 

```python
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor(random_state=1)
```

```python
#Fit training data into model
dt_model.fit(X_train, Y_train)
```

```python
Y_preds = dt_model.predict(X_valid.head())
```

```python
Y_preds
```

```python
pd.DataFrame({'Y' : Y_valid.head(),'Y_preds' : Y_preds})
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor(random_state=1)
rf_model.fit(X_train,Y_train)
```

```python
rf_val_preds = rf_model.predict(X_valid)
```
